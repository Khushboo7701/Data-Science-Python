```python
'Matplotlib is a low level graph plotting library in python that serves as a visualization utility.'
'Install it using the command: pip install matplotlib'
import matplotlib.pyplot as plt
import numpy as np

#axes and ticklabels
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'ys-') # yellow coloured solid line with square marker 
l2 = ax.plot(x2,y,'go--') # green coloured dashed line with circle marker
ax.legend(labels = ('tv', 'Smartphone'), loc = 'lower right') # legend placed at lower right
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
ax.set_xticks([10,30,50,70])
ax.set_xticklabels(['Very Less','Low','Average','High'])

plt.show()

from mpl_toolkits.axisartist.axislines import Subplot
#axis and figure
fig = plt.figure(figsize =(4, 4))

ax = Subplot(fig, 111)
fig.add_subplot(ax)

ax.axis["left"].set_visible(False)
ax.axis["bottom"].set_visible(False)

ax.set_title('Figure',
          fontsize = 14, fontweight ='bold')
plt.show()


#Important parts of a matplotlib figure
'''Major tick, minor tick, major tick label, minor tick label, y-axis label,
x-axis label, figrue, axis, axes, plot, title, legend, grid, markers, spines
'''
```

```python
#line plots
#plotting line-multiple points
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()

#plotting line with default xpoints- 0,1,2,3, and so on
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints)
plt.show()

#plotting only points
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()

'''
format strings fmt - marker|line|color
Marker: emphasize each point with a specified marker
line reference: type of line(dotted, solid, dashed, etc)
color reference: color of line
'''
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r') #prints a red dotted line with ends marker as filled rings
plt.show()

plt.plot(ypoints, '*-c') #prints a cyan solid line with ends marker as *
plt.show()

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker='o', ms=20)
plt.show()
```

```python
# Stackplot with X, Y, colors value
days = [1, 2, 3, 4, 5]
Studying = [7, 8, 6, 11, 7]
playing =  [8, 5, 7, 8, 13]
plt.stackplot(days, Studying, playing, colors =['r', 'c'])

plt.xlabel('Days')
plt.ylabel('No of Hours')
plt.title('Representation of Study and Playing wrt to Days')
plt.show()
```

```python
plt.axes(projection = 'polar')
  
# setting the radius
r = 2
  
# creating an array containing the radian values
rads = np.arange(0, (2 * np.pi), 0.01)
  
# plotting the circle
for rad in rads:
    plt.polar(rad, r, 'g.')

plt.show()
```

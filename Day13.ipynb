```python
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.get_dataset_names())

data = sns.load_dataset('iris')
```

```python
print(data.head(5))
print(data.tail(5))
print(data.describe(include="all"))
sns.lineplot(x="sepal_length", y="sepal_width",data=data)
plt.title("Plotting using Matplotlib function")
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.barplot(x="species",y="sepal_length",data=data)
```

```python
sns.barplot(x="species",y="sepal_length",data=data, hue='sepal_width', order=['virginica','setosa','versicolor'])
```

```python
sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
sns.countplot(x='species',data=data)
plt.show()
```

```python
sns.set_theme(style='ticks',color_codes=True)
sns.stripplot(x ='species', y ='sepal_length', data = data,jitter = False)
```

```python
sns.swarmplot(x='species',y='sepal_width',data=data)
plt.show()
```

```python
sns.pointplot(x='species',y='sepal_width',data=data)
plt.show()
```

```python
sns.factorplot(x ='species', y ='sepal_length', data = data,kind='bar')
```

```python
tc = data.corr()
print(tc)
sns.heatmap(tc)
plt.show()
```

```python
sns.distplot(data['sepal_length'], kde = False, color ='hotpink', bins = 30)
plt.show()
```

```python
sns.jointplot(x ='sepal_length', y ='sepal_width', data = data, kind ='kde')
plt.show()
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
g = sns.PairGrid(data=data)
g.map(plt.scatter) #All the grids plot scatter plot
plt.show()
```

```python
#Simple Linear plot-
sns.set_style('whitegrid')
sns.lmplot(x='sepal_length',y='sepal_width',data=data)
plt.show()

sns.lmplot(x='sepal_length',y='petal_length', data=data)
plt.show()
```

```python
#linear Plot with additional parameters
sns.lmplot(x='sepal_length',y='sepal_width',data=data,hue='species',markers=['o','v','s'])
plt.show()
```

```python
sns.set_style('dark') #set_style is used for plot styling

sns.countplot(x='species',data=data,palette = 'bright')
#The despine() is a function that removes the spines from the right and upper portion 
#of the plot by default
sns.despine()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='species',data=data,palette = 'dark')
plt.show()
```

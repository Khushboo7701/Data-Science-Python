```python
import numpy as np
import pandas as pd
import sys
df = pd.read_csv('PS4_GamesSales.csv')
df.dropna(0,inplace = True)
df
```

```python
print(np.mean(df))
```

```python
#We can also mention the axis along which the mean can be calculated.
print("Mean along axis 0: ", np.mean(df,axis=0))
print("Mean along axis 1: ", np.mean(df,axis=1))
```

```python
print(np.shape(df))
print("Minimum along axis 1: ", np.min(df,axis=1))
print("Minimum along axis 0: ", np.min(df,axis=0))
print("Maximum along axis 0: ", np.max(df,axis=0))
print("Maximum along axis 1: ", np.max(df,axis=1))
print("Standard deviation: ", np.std(df))
print("Variance: ", np.var(df,axis=1))
print("Average: ", np.average(df['Year']))
print("Percentile:", np.percentile(df['Year'],50))
```

```python
print("Range : ", np.ptp(df['Year']))
print("Range : ", np.ptp(df['Global']))
```
